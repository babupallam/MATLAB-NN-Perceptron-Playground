## Aim of the Paper:
The paper aims to generalize the Recursive Deterministic Perceptron (RDP) model from 2-class classification problems to **m-class problems**. The primary objective is to adapt the RDP's deterministic approach to handle multiple classes by introducing a new notion of **linear separability**.

## Key Proposals:
1. **Generalization of Linear Separability**:
   - The paper extends the concept of linear separability from two classes to **m classes**.
   - The authors introduce a method where **m classes are linearly separable by translating a single hyperplane m+1 times**, each translation using the same vector.

2. **Transformation from m-Class to 2-Class Problems**:
   - The proposed approach reduces the complexity of m-class problems by transforming them into multiple 2-class problems.
   - The authors present a specialized algorithm that transforms non-linearly separable (NLS) m-class problems into linearly separable (LS) ones.

3. **m-RDP Model**:
   - The **m-RDP** model is introduced, which extends the 2-class RDP to m classes.
   - The model uses a sequence of intermediate neurons (INs) to create a layered topology that separates m classes by translating hyperplanes.

## Example:
- The paper provides an example of a 3-class classification problem. The example demonstrates how the proposed method builds a 3-RDP that linearly separates the three classes using transformed input vectors and hyperplane translations.

## Conclusion:
- The paper concludes by proposing that the new notion of linear separability for m classes is effective for transforming complex classification problems into simpler ones. The m-RDP model can be used to approximate continuous functions with finite domains, making it versatile for various applications.

## **3. Specialized NLS to LS Transformation Algorithm for m Classes**

This section introduces the specialized algorithm that transforms non-linearly separable (NLS) m-class classification problems into linearly separable (LS) ones, building on the ideas presented in the previous sections. The algorithm adapts the 2-class transformation algorithm to handle m classes using the new notion of linear separability.



### **3.1 Overview of the Algorithm**

- The goal of the algorithm is to take an m-class NLS problem and iteratively transform it into a linearly separable (LS) one.
- The process works by:
  - Selecting a linearly separable (LS) subset that belongs to a single class.
  - Adding an **artificial variable** to the input vectors to aid in the separation.
  - Repeating the process until all classes become linearly separable.



### **3.2 Steps of the Algorithm**

- **Step 1: Initialization**
  - Start with m non-linearly separable classes \( X_1, X_2, ..., X_m \).
  - Initialize the sequence of augmented input vectors for each class.
  
- **Step 2: Selection of LS Subset**
  - At each iteration, select a non-empty subset from one of the m classes that is linearly separable from the other classes.
  - This subset must belong to a single class, and its separation is performed by introducing an artificial variable.

- **Step 3: Addition of Artificial Variable**
  - An artificial variable is added to the input vectors, and the values of this variable are assigned based on the class membership.
  - If the selected LS subset belongs to class \( j \), assign a value \( b_j \) to the input vectors of the subset and \( b_{j+1} \) to the rest of the input vectors.
  - If the subset belongs to the last class, assign a value \( b_m \) to the subset and \( b_{m-1} \) to the rest.

- **Step 4: Repeated Transformation**
  - The process continues iteratively, selecting new LS subsets and adding artificial variables until all classes become linearly separable.
  
- **Termination**: The algorithm terminates once all classes are separated by a hyperplane.
