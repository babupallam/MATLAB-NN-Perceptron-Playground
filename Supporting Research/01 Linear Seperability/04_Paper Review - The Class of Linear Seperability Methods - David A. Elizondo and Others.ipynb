## Title: **The Class of Linear Separability Method**

## Author:
Presented at the **ESANN 1997 proceedings** - European Symposium on Artificial Neural Networks, held in Bruges, Belgium (16-18 April 1997).

## Aim of the Paper:
The aim of the paper is to introduce and evaluate the **Class of Linear Separability (CLS) Method** for solving classification problems. This method is used to identify linearly separable (LS) subsets within non-linearly separable (NLS) datasets, providing a more efficient way to handle complex data classification tasks.

## Problem Addressed:
- **Linear Separability**: In machine learning and neural networks, many problems involve identifying whether two classes can be separated by a hyperplane in a given feature space.
- **NLS Problems**: Non-linearly separable datasets present a challenge, as traditional perceptron algorithms or hyperplane-based classification models fail to separate the classes.

## Proposed Method:
- The **Class of Linear Separability (CLS) Method** is proposed as a solution for identifying hyperplanes that separate LS subsets within an NLS dataset.
- This method systematically evaluates points in the dataset to find hyperplanes that maximize the number of linearly separable points, allowing for better classification even when the entire dataset is not linearly separable.

## Key Contributions:
1. **Efficiency**: The CLS method is computationally efficient compared to exhaustive methods that test all possible hyperplanes. It tracks hyperplanes that separate the largest subsets without recalculating for each new point.
2. **Subset Identification**: It identifies linearly separable subsets, which can be used to simplify neural network architectures, such as in Recursive Deterministic Perceptron (RDP) networks.

## Application:
- The method can be applied to real-world classification tasks, improving the performance of neural networks and other machine learning models by handling non-linearly separable data more effectively.

## Conclusion:
The CLS method presents a significant advancement in identifying LS subsets within NLS datasets, offering an efficient solution to improve classification tasks in complex datasets.

### **1. Introduction**



#### **A. Importance of Linear Separability (LS)**
- In two-class classification problems, knowing whether the classes are linearly separable (LS) can greatly simplify the neural network's topology.
  - When the data is LS, a **single-layer perceptron** can classify the data without needing complex architectures.
  - LS problems allow for efficient classification and minimal network complexity.



#### **B. Recursive Deterministic Perceptron (RDP)**
- For non-linearly separable (NLS) problems, identifying **LS subsets** can help recursively build a network topology.
  - This approach is similar to how the **Recursive Deterministic Perceptron (RDP)** is constructed, where subsets of LS points are used to reduce the complexity of the network.
  - The RDP is a network structure that uses recursive methods to transform NLS data into LS subsets for classification purposes.
