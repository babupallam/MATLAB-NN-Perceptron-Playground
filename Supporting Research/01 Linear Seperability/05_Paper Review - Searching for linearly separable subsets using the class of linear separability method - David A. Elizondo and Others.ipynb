# Paper Summary

## Title: **Searching for Linearly Separable Subsets Using the Class of Linear Separability Method**

## Author:
- **David Elizondo**

## Aim of the Paper:
The aim of this paper is to propose a more efficient method to find linearly separable (LS) subsets from non-linearly separable (NLS) datasets. The focus is on identifying LS subsets of maximum cardinality, which can reduce the complexity of neural network architectures, particularly in Recursive Deterministic Perceptron (RDP) networks.

## Problem Addressed:
- **Max-Separability Problem**: Finding LS subsets of maximum cardinality in NLS classification problems is an NP-complete problem. This makes it computationally expensive and time-consuming using exhaustive strategies.
  
- **RDP Neural Networks**: LS subsets are crucial for constructing RDP neural networks. This paper seeks to simplify the process by using the Class of Linear Separability (CLS) method.

## Proposed Method:
- The **Class of Linear Separability (CLS) Method** is used to test linear separability and to find LS subsets directly within NLS datasets. This method avoids the exhaustive strategies typically used in this process.
  
- The CLS method characterizes hyperplanes (HPs) that can linearly separate subsets of data points in complex classification problems.

## Key Contributions:
1. **Efficient Subset Identification**: The CLS method finds LS subsets more efficiently compared to traditional exhaustive methods by focusing on hyperplanes that separate the largest number of points without additional recalculation.
  
2. **Application to Neural Networks**: The identified LS subsets can be used to reduce the size and complexity of RDP neural network topologies.

## Application:
- The paper applies the CLS method to a 2-class 2D classification problem and demonstrates how the method can find hyperplanes that linearly separate subsets of points within a non-linearly separable problem.

## Conclusion:
The CLS method is shown to be a direct and efficient approach for finding LS subsets in NLS problems. It avoids the computational burden of exhaustive methods and enables the construction of optimized neural networks with fewer computational resources.


# Detailed Review

### **1. Introduction**

#### **A. Overview of the Linear Separability Problem**
- **Linear Separability in Classification**:
  - The paper addresses the problem of determining whether two classes in a two-class classification problem are **linearly separable (LS)**.
  - In the simplest case, a **single-layer perceptron** network can classify LS problems efficiently.

- **Non-Linearly Separable (NLS) Problems**:
  - When classes are **non-linearly separable (NLS)**, the task becomes more complex. However, even in NLS problems, there can be subsets of points that are LS.
  - These LS subsets can be used to simplify the topology of neural networks, specifically the **Recursive Deterministic Perceptron (RDP)**.

#### **B. Importance of LS Subsets**
- **LS Subsets for Neural Networks**:
  - LS subsets, especially those with maximum cardinality, are of particular interest because they minimize the size of the neural network topology.
  - These subsets can be used to build an efficient network structure like the RDP, which adds intermediate neurons to solve the NLS problem.

- **Max-Separability Problem**:
  - Finding the LS subset with the highest cardinality is known as the **Max-Separability Problem**.
  - This is an **NP-complete** problem, meaning it is computationally intensive, especially as the dimensionality of the data increases.
  - For fixed dimensions, a polynomial-time algorithm exists, but it is still resource-heavy.

#### **C. Traditional Methods and Their Limitations**
- **Exhaustive Search Strategy**:
  - The traditional approach for finding LS subsets involves an **exhaustive search**, which compares every point to every other point.
  - The method keeps track of subsets, testing linear separability at each step and discarding points that make the subset non-separable.
  - This exhaustive method is computationally expensive and impractical for large datasets.

- **Alternative Approaches**:
  - Other methods, such as **modifications of the perceptron algorithm**, have been proposed to handle LS subsets more efficiently. These modifications aim to reduce computational complexity by using discrete data inputs or restricting inputs to specific ranges (e.g., binary or real-valued vectors).

#### **D. Proposed Approach**
- **Class of Linear Separability (CLS) Method**:
  - The paper introduces a novel method based on the **Class of Linear Separability (CLS)**, which provides a more efficient solution for finding LS subsets.
  - The CLS method directly tests linear separability without the need for exhaustive recalculations, making it computationally less expensive.
  - This method visits a finite set of hyperplanes (HPs) and tests whether they linearly separate the points in the dataset.

#### **E. Structure of the Paper**
- The paper is divided into five sections:
  1. **Introduction**: Discusses the problem of linear separability and the importance of LS subsets in neural network topology.
  2. **Background**: Provides standard definitions and notations.
  3. **CLS Method**: Explains the Class of Linear Separability (CLS) method for testing linear separability.
  4. **Application**: Shows how the CLS method can be used to find LS subsets within NLS problems.
  5. **Conclusion**: Summarizes the results and suggests directions for future research.

### **4. Application**



#### **A. Using CLS Method to Find Linearly Separable Subsets**
- **General Concept**:
  - The CLS method, as previously introduced, is used to determine whether two sets of data are linearly separable (LS).
  - If the two sets are LS, the method finds a hyperplane (HP) that separates them. However, if the sets are not linearly separable, the CLS method can still be applied to identify subsets of points that are linearly separable from the rest of the dataset.

- **Searching for LS Subsets**:
  - The CLS method searches within a **finite set of hyperplanes** to find ones that separate the two classes, \( X \) and \( Y \), based on certain conditions.
    - These hyperplanes must pass through at least one point from each class.
    - Some hyperplanes may not fully separate the two classes but will separate **subsets** of points from one class from the rest of the data.

- **Efficiency**:
  - Unlike exhaustive search methods, the CLS method tracks hyperplanes that linearly separate subsets of points without requiring additional computations.
  - This results in a more efficient identification of the linearly separable subsets (LS subsets), particularly those with **maximum cardinality** (the largest possible number of points in the subset).
