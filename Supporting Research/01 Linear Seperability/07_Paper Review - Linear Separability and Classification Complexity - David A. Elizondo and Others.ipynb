## Title **Linear Separability and Classification Complexity**

## Authors
- **David A. Elizondo**, Ralph Birkenhead, Matias Gamez, Noelia Garcia, Esteban Alfaro

## Aim of the Paper
The aim of the paper is to explore the relationship between linear separability and classification complexity. It investigates how linearly separable problems are easier to solve and proposes a method to quantify classification complexity by reducing non-linearly separable problems to sequences of linearly separable steps. This method is intended to measure the complexity of classification datasets.

## Key Proposals
- **Quantifying Complexity**: The paper introduces a method that reduces any two-class classification problem to a sequence of linearly separable steps. The number of these steps serves as a measure of the problem's complexity.
- **Recursive Deterministic Perceptron (RDP)**: A neural network approach is used, where the complexity of the network topology reflects the difficulty of the data set.
- **Complexity Measure**: The proposed complexity measure is based on the number of intermediate neurons required to make a non-linearly separable problem linearly separable.

## Methods
- **Linear Separability Testing**: The paper tests various methods for determining whether two data classes are linearly separable, including algorithms such as Simplex, convex hull methods, and perceptron learning.
- **RDP Neural Networks**: The paper builds RDP models using batch, incremental, and modular learning methods to address non-linearly separable problems.
  
## Benchmarks
The study applied the proposed complexity measure on several benchmarks, including:
- Iris
- Soybean
- Wisconsin Breast Cancer
- Glass
- Hepatitis
- Ionosphere
- Sonar
- Wine
- MONK’s Problem

## Conclusion
The proposed complexity measure effectively quantifies the difficulty of classification problems. The study finds that linearly separable problems are simpler and that the number of steps required to achieve separability serves as a strong indicator of classification complexity.

### 1. Introduction

#### Overview
- **Focus**: The paper investigates the impact of different linear separability testing methods on the performance of constructive neural network algorithms, particularly the Recursive Deterministic Perceptron (RDP).
- **Goal**: To understand how these testing methods affect the topology size, convergence time, and generalization level of neural networks.

#### Recursive Deterministic Perceptron (RDP)
- **Description**:
  - A neural network algorithm that can deterministically separate two or more classes, even if they are not linearly separable using a simple perceptron.
  - The RDP augments the affine dimension of the input vector by adding outputs from intermediate neurons (each corresponding to a single-layer perceptron) as new components.
  - The result is a multi-layer topology capable of solving non-linearly separable (NLS) classification problems.

#### Testing Methods for Linear Separability
- **Approaches**:
  - The paper explores multiple methods for testing linear separability, which are essential for transforming NLS problems into linearly separable (LS) problems.
  - It contrasts the impact of exact methods like the Simplex algorithm, Convex Hull, Support Vector Machines (SVM), and Perceptron, as well as approximative methods like Anderson and Fisher.

#### Key Objectives
- **Main Aim**: The study focuses on empirically testing how these separability methods influence the performance of RDP neural networks, considering:
  - Topology size
  - Convergence time
  - Generalization ability

### 2. Background

#### 2.1 Introduction to RDP Neural Network
- **Recursive Deterministic Perceptron (RDP)**:
  - A multilayer feedforward neural network designed to address any two-class classification problem.
  - Works effectively for non-linearly separable (NLS) problems by converting them into a series of linearly separable (LS) problems.
  
#### 2.2 Transformation of Non-Linearly Separable Problems
- **Process**:
  - The RDP transforms a non-linearly separable problem into multiple LS problems by progressively adding intermediate neurons.
  - These neurons act as simple perceptrons, separating subsets of linearly separable points from the rest of the data.
  - The neural network builds incrementally, allowing it to solve NLS classification tasks.
  
- **Example**:
  - The paper illustrates the process using the classic XOR problem, which is a 2-input NLS problem.
  - Intermediate neurons are used to transform this NLS problem into LS problems.

#### 2.3 Key Features of RDP
- **Guaranteed Convergence**:
  - The RDP algorithm is guaranteed to find a solution to any two-class classification problem, even when the data is not linearly separable.
  - The algorithm incrementally adds intermediate neurons until the classes become linearly separable.
  
- **No Catastrophic Interference**:
  - Unlike other neural networks, such as those trained with backpropagation, the RDP does not suffer from catastrophic interference (i.e., losing previously learned information when learning new data).

- **Transparency of Knowledge Extraction**:
  - The RDP provides a clear and interpretable set of rules that can be extracted using computational geometry approaches.
  - These rules are represented as a union of open polyhedral sets, allowing easy interpretation of the network's decision-making process.

### 3. Methods for Testing Linear Separability

#### 3.1 Overview of Linear Separability

- **Definition**:
  - Two subsets \( X \) and \( Y \) of \( R^d \) are said to be linearly separable (LS) if there exists a hyperplane \( P \) such that the elements of \( X \) and those of \( Y \) lie on opposite sides of it.
  
#### 3.2 Classification of Methods

- **Six Methods Used in the Study**:
  - The study uses four **exact methods** and two **approximative methods** for testing LS:
    - **Exact Methods**:
      1. **Simplex Algorithm**: Solves constrained linear equations to check if two classes are LS.
      2. **Convex Hull Algorithm**: Tests if the convex hulls of two classes intersect.
      3. **Perceptron Algorithm**: Ensures convergence for LS classes.
      4. **Support Vector Machine (SVM)**: Uses quadratic programming to find the optimal separating hyperplane.
    - **Approximative Methods**:
      1. **Anderson’s Method**: Assumes a multivariate Gaussian distribution to find a separating hyperplane.
      2. **Fisher Linear Discriminant**: Finds a linear combination of input variables to separate classes.

#### 3.3 Exact Methods

- **Simplex Algorithm**:
  - A classic method for solving linear programming problems by optimizing an objective function subject to constraints.
  - **Outputs**:
    - If a solution exists, the algorithm finds it.
    - The model may be solvable, infeasible, or unbounded.

- **Convex Hull Algorithm**:
  - LS occurs if the convex hulls of two sets do not intersect.
  - Utilizes the Kozinec's algorithm to build vectors that converge into a final separating hyperplane.
  - **Outputs**:
    - Always finds a solution for LS sets.
  
- **Perceptron Algorithm**:
  - Ensures convergence in a finite number of steps for LS problems.
  - **Outputs**:
    - Identifies a separating hyperplane for LS classes.

- **Support Vector Machine (SVM)**:
  - Uses quadratic programming to find the optimal hyperplane that maximizes the margin between classes.
  - **Outputs**:
    - Offers robust performance but requires fine-tuning of parameters.

#### 3.4 Approximative Methods

- **Anderson’s Method**:
  - Relies on assumptions of Gaussian distribution for classes.
  - Minimizes misclassification errors when separating non-linearly separable (NLS) classes.

- **Fisher Linear Discriminant**:
  - Focuses on maximizing the separation of the class projections while minimizing within-class variance.
  - Provides an approximate solution without assuming a probability distribution.


### 4. Methods for Building RDP Neural Networks

#### 4.1 Overview of Methods

- **Three Methods for Building RDP Networks**:
  - **Batch Learning**:
    - This method focuses on selecting homogeneous linearly separable (LS) subsets from a non-linearly separable (NLS) set.
    - It uses an NP-complete strategy, making it computationally expensive but effective in generating small topologies.
  
  - **Incremental Learning**:
    - The method used in this study.
    - It allows for new knowledge to be incorporated without rediscovering previously learned information.
    - Has a polynomial time complexity of **O(n log n)**, ensuring computational efficiency while maintaining accuracy.
    
  - **Modular Learning**:
    - This method combines several RDP models without the need for further training.
    - Its computational complexity is **O(n²)**, which is higher compared to the Incremental approach.

#### 4.2 Description of the Incremental Method

- **Progressive Learning Process**:
  - The Incremental method begins with training an RDP network using a subset of the training data.
  - Initially, a subset of points from the class with maximum cardinality is used for training.
  - The trained network is then extended with data points from other classes.
  - New points are passed through the existing RDP network, and when a misclassification occurs, an intermediate neuron (IN) is added to correct the error without disturbing previously learned data.

#### 4.3 Example of the Incremental Method

- **Illustrative Case**:
  - The process is illustrated using a simple two-dimensional classification problem.
  - A subset of data points from two classes is selected, and the network incrementally adds neurons to create a separating hyperplane for each new set of points.
  - This example demonstrates how the method adds complexity to the neural network progressively to achieve correct classification.

### 5. Quantification of the Complexity of Classification Problems



#### 5.1 Overview
- **Objective**:
  - The complexity of classification problems is measured based on the degree of linear separability between the classes in the dataset.
  - This method focuses on transforming non-linearly separable (NLS) problems into linearly separable (LS) ones and uses the number of intermediate steps (hyperplanes) required for this transformation as a measure of complexity.
  


#### 5.2 Two Cases of Complexity

- **Case 1: Linearly Separable (LS) Classes**:
  - If the dataset is LS, the problem can be solved using a single hyperplane that separates the two classes.
  - **Complexity level**: The simplest case, represented by a complexity value of **0** (Fig. 6 shows a two-class LS problem).
  
- **Case 2: Non-Linearly Separable (NLS) Classes**:

  - **Complexity level**: In the worst-case scenario, the complexity level approaches **1** (Fig. 7 shows an NLS problem requiring multiple hyperplanes).

### 7. Results and Conclusions

#### 7.1 Overview of Results

- **Objective**:
  - The section provides a performance comparison of Recursive Deterministic Perceptron (RDP) neural networks using six different linear separability testing methods.
  - Performance is measured in terms of:
    - **Convergence Time**
    - **Generalization Capability**
    - **Topology Size** (number of intermediate neurons required).

#### 7.2 Performance of Exact Methods

- **Exact Methods**:
  - The exact methods used were Simplex, Convex Hull, Support Vector Machine (SVM), and Perceptron.
  - **Convergence Time**:
    - **RDP-Simplex** performed the best, consistently achieving the fastest convergence times.
    - **RDP-SVM** showed comparable performance but was slightly slower than Simplex in many cases.
  - **Topology Size**:
    - **RDP-Simplex** produced the smallest topologies across the majority of datasets, minimizing the number of intermediate neurons required.
    - **RDP-SVM** came close to Simplex in terms of topology size but did not outperform it.
  - **Generalization**:
    - No clear winner emerged in terms of generalization.
    - Each method had a dataset where it underperformed, achieving poor results on specific datasets like Glass, where generalization fell below acceptable levels.

#### 7.3 Performance of Approximative Methods

- **Approximative Methods**:
  - The approximative methods used were Anderson and Fisher.
  - **Convergence Time and Topology Size**:
    - **RDP-Anderson** consistently outperformed **RDP-Fisher** in both convergence time and topology size.
  - **Generalization**:
    - As expected, approximative methods did not achieve generalization results as high as exact methods.
    - Anderson was better than Fisher in generalization, but still fell short compared to Simplex and SVM.

#### 7.4 Comparative Performance

- **Exact vs. Approximative Methods**:
  - Exact methods generally outperformed approximative methods across all metrics.
  - The study finds no compelling reason to prefer approximative methods like Anderson and Fisher, as they did not offer significant advantages in terms of speed, topology size, or generalization.

#### 7.5 Conclusions

- **RDP-Simplex**:
  - Outperformed other methods in convergence time and topology size.
  - Provided stable performance without requiring fine-tuning or parameter adjustment, making it robust and efficient.
- **RDP-SVM**:
  - Showed comparable results to Simplex in terms of topology size but required more computational time and parameter adjustments.
- **Final Recommendation**:
  - Exact methods like Simplex and SVM are preferred over approximative methods for constructing RDP neural networks due to their superior performance in most benchmarks.
  
- **Future Work**:
  - The study suggests further exploration into optimizing approximative methods, although the exact methods currently dominate in efficiency and stability.
