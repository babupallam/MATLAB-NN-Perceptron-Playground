## Title: **Searching for Linearly Separable Subsets Using the Class of Linear Separability Method**

## Author:
- **David Elizondo**
  - School of Computing, De Montfort University, Leicester, UK.

## Aim of the Paper:
 - The paper aims to propose a novel and more efficient method to find linearly separable (LS) subsets from non-linearly separable (NLS) datasets using the Class of Linear Separability (CLS) method.
 - It focuses on finding LS subsets of maximum cardinality, which are critical for minimizing neural network topology, especially in Recursive Deterministic Perceptron (RDP) networks.

## Problem Addressed:
- **Max-Separability Problem**: Finding the largest LS subset from an NLS dataset is an NP-Complete problem.
- Traditional exhaustive methods for finding LS subsets are computationally expensive.

## Proposed Solution:
- The paper introduces the **CLS method** to directly find LS subsets from within NLS datasets.
- The CLS method identifies all hyperplanes (HPs) that pass through one point from each class and tests them to find LS subsets.
- By keeping track of the hyperplanes that linearly separate the largest number of points, LS subsets of maximum cardinality are found without extra computation.

## Key Contributions:
1. **CLS Method**:
   - The CLS method computes hyperplanes that separate two classes and identifies subsets of LS points from the rest of the dataset.
   - It avoids the need for weight adjustments as required in other methods like the perceptron algorithm.

2. **Application to RDP Neural Networks**:
   - The LS subsets found using the CLS method can be used to build efficient neural network topologies (RDP neural networks).
   - These subsets reduce the complexity of the network's architecture.

## Main Results:
- The proposed method is computationally efficient compared to exhaustive approaches.
- It avoids expensive recalculations by directly visiting all relevant hyperplanes that can separate LS subsets.

## Future Directions:
- The paper suggests further research on optimizing the CLS method for even larger and more complex datasets.

# Detailed Review




## 1. Introduction

### A. Problem Overview
- **Non-Linearly Separable (NLS) Problems**:
  - In two-class classification problems where the data is non-linearly separable (NLS), at least one point can always be found that is linearly separable (LS) from the rest of the points.
  - This is crucial for constructing Recursive Deterministic Perceptron (RDP) neural networks, which rely on LS subsets to reduce the neural network's topology.

- **Max-Separability Problem**:
  - The challenge of finding the LS subset with the highest cardinality (the largest number of LS points) is known as the **Max-Separability Problem**.
  - This is an NP-Complete problem, meaning that it is computationally difficult and time-consuming to solve.

### B. Relevance to Neural Networks
- **Recursive Deterministic Perceptron (RDP) Networks**:
  - RDP networks can be constructed by finding LS subsets from NLS data.
  - These networks operate by adding **intermediate neurons (INs)**, which augment the affine dimension of the input vectors, transforming NLS problems into LS problems.
  
- **Traditional Methods**:
  - Traditional methods for finding LS subsets involve exhaustive testing, which checks every point against the others.
  - This is computationally expensive because it requires running LS tests repeatedly as new points are added to the subset.

### C. Goal of the Paper
- **Objective**:
  - The paper proposes a novel, more efficient method for finding LS subsets of maximum cardinality, based on the **Class of Linear Separability (CLS) Method**.
  
- **Proposed Method**:
  - The CLS method tests linear separability directly, providing a more direct and efficient way of finding LS subsets compared to exhaustive methods.

- The paper is divided into the following sections:
  1. **Introduction**: Overview of the problem and proposed solution.
  2. **Background**: Notations and definitions relevant to LS and NLS classification problems.
  3. **CLS Method**: Explanation of the CLS method used for testing linear separability.
  4. **Application of CLS to LS Subsets**: Detailed method for applying CLS to find LS subsets within NLS datasets.
  5. **Conclusions**: Summary of the findings and potential future directions.

## 3. Methods for Testing Linear Separability

### A. Convex Hull Method
- **Convex Set**:
  - A set is convex if, for any two points within the set, the line segment joining these two points lies entirely within the set.
  
- **Convex Hull**:
  - The convex hull of a set \( S \) is the smallest convex set containing all the points in \( S \).
  
- **Testing Linear Separability**:
  - Two sets \( X \) and \( Y \) are linearly separable if and only if the convex hulls of \( X \) and \( Y \) do not intersect.
  
- **Algorithm**:
  - A convex hull-based algorithm is proposed for testing linear separability, which constructs the convex hulls for both sets and checks if their intersection is empty.
  
- **Complexity**:
  - Convex hull methods work efficiently in lower-dimensional spaces but become computationally intensive in higher dimensions.

### B. Halting Upper Bound for the Perceptron
- **Perceptron Learning Algorithm**:
  - The perceptron learning algorithm can be used to classify linearly separable datasets. It converges in a finite number of steps for linearly separable sets.
  
- **Upper Bound**:
  - A computable halting upper bound, \( K(X, Y) \), is proposed for the perceptron algorithm. If the number of steps exceeds this bound, the two sets are not linearly separable.
  
- **Impact**:
  - This bound prevents infinite loops in cases where the dataset is non-linearly separable.

### C. Class of Linear Separability Method
- **Characterization of Points**:
  - The Class of Linear Separability (CLS) method focuses on identifying a hyperplane that linearly separates two classes \( X \) and \( Y \).
  
- **Algorithm**:
  - The CLS method characterizes a set of points that defines a hyperplane \( P_Z \) and recursively tests for separability. If a separating hyperplane is found, it confirms that the sets are linearly separable.
  
- **Recursive Dimensional Reduction**:
  - The CLS method works by transforming the separability problem from higher dimensions to lower dimensions and recursively checks for separability at each level.
  
- **Example**:
  - A detailed example is provided where the method is applied to a 2-class problem in \( \\mathbb{R}^2 \), demonstrating how the hyperplane is identified and used to separate the classes.

### D. Complexity
- The complexity of the CLS method is discussed, noting that while it can be computationally expensive, it is versatile in identifying linearly separable subsets within a non-linearly separable dataset.

## 4. Application of CLS Method to LS Subsets

### A. Purpose of the CLS Method in LS Subsets
- The CLS method is designed not only to test whether two sets are linearly separable (LS) but also to identify **LS subsets** within a **non-linearly separable (NLS)** dataset.
  
- The key idea is that subsets of points from an NLS set may still be linearly separable from the rest, and the CLS method can help find those subsets with maximum cardinality (the largest number of points possible that are LS).

### B. Searching for Hyperplanes (HPs)
- The CLS method searches through a finite number of hyperplanes that pass through at least one point of each class.
  
- **Conditions for HPs**:
  - The HP must pass through one point from each of the two classes.
  - The HP must separate the points of one class from the other as linearly as possible.

### C. Identifying LS Subsets
- While the primary objective of the CLS method is to find a hyperplane that linearly separates two sets, some hyperplanes may not fully separate the classes.
  
- These HPs, however, may linearly separate **subsets** of points from one class from the rest of the points in the dataset.
  
- The method is efficient in tracking which hyperplanes separate the largest subsets of points without the need for excessive recalculation.

## 5. Conclusion

### A. Summary of Key Contributions
- The paper presents three methods for testing linear separability:
  1. **Convex Hull Method**:
     - This method checks whether the convex hulls of two sets intersect. If they don't intersect, the two sets are linearly separable.
     
  2. **Perceptron Halting Upper Bound**:
     - The authors propose an upper bound on the perceptron algorithm. If the number of steps taken by the perceptron exceeds this bound, it indicates that the sets are not linearly separable.

  3. **Class of Linear Separability (CLS) Method**:
     - This method focuses on finding hyperplanes that separate linearly separable subsets within non-linearly separable datasets.

### B. Implications for NLS Classification Problems
- **Approximation Approach**:
  - The authors address non-linearly separable (NLS) problems by approximating them with a linearly separable (LS) problem. They note that finding the best approximation is an NP-complete problem.
  
- **Recursive Deterministic Perceptron (RDP)**:
  - The Recursive Deterministic Perceptron (RDP) model is introduced to handle NLS classification by transforming NLS problems into LS ones by adding intermediate neurons.
  - This method guarantees a solution within a limited number of steps.

### C. Main Results
- **Computational Efficiency**:
  - The CLS method and halting upper bound for perceptrons improve the efficiency of determining linear separability.
  - These methods provide computationally feasible solutions, especially for high-dimensional or complex classification tasks.

### D. Future Directions
- **Hyperplane Optimization**:
  - Further research should focus on optimizing hyperplanes that maximize generalization when separating two linearly separable classes.
  
- **Probabilistic Linear Separability**:
  - Investigating probabilistic linear separability could help make classification decisions more flexible, particularly for datasets that are not perfectly linearly separable.
  
- **Multi-Class Classification**:
  - Extending these methods to handle multi-class classification problems and larger datasets is a key future direction for this research.
