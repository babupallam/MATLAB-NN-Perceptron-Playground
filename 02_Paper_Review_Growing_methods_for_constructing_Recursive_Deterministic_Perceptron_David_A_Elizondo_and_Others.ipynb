**Title**: Growing Methods for Constructing Recursive Deterministic Perceptron Neural Networks and Knowledge Extraction

**Authors**: M. Tajine and D. Elizondo

**Published in**: Artificial Intelligence, 1998

## Aim of the Paper:
The paper aims to introduce the **Recursive Deterministic Perceptron (RDP)**, a multilayer neural network model, as a generalization of the Single Layer Perceptron Topology (SLPT). The RDP overcomes the limitations of SLPT by solving any two-class classification problem, including those involving non-linearly separable (NLS) data.

## Key Proposals:
1. **Recursive Deterministic Perceptron (RDP)**: A model capable of handling both linearly separable and non-linearly separable classification problems by augmenting input vectors with outputs from intermediate neurons (INs).
2. **Growing Methods**: Three different methods for constructing RDPs are introduced:
   - **Batch Learning**: Selects linearly separable subsets in a batch process to build RDPs.
   - **Incremental Learning**: Dynamically updates RDPs with new knowledge without disturbing previously acquired information.
   - **Modular Learning**: Decomposes complex problems into smaller, independently solvable sub-problems and assembles them into a complete RDP.

## Knowledge Extraction:
- The paper discusses how knowledge embedded in RDPs can be extracted and expressed as a union of decision regions (open polytopes), offering interpretability over the "black-box" nature of other neural networks like backpropagation.
- **Boolean Operations**: Methods to combine decision regions of RDPs using Boolean operations are also introduced.

## Results:
- The RDP model was tested across several classification problems and compared with other models such as Backpropagation (BP) and Cascade Correlation (CC).
- The results showed that the RDP guarantees convergence and offers advantages such as self-determined topology, ease of rule extraction, and avoidance of catastrophic interference.

**Keywords**: Batch learning, Incremental learning, Modular learning, Recursive Deterministic Perceptron, Knowledge extraction, Boolean operations, Linear separability.


# Detailed Review


**1. Introduction**

**1.1. Background on Single Layer Perceptron (SLPT)**
- The **Single Layer Perceptron Topology (SLPT)** was introduced by Rosenblatt and is limited to solving linearly separable (LS) classification problems.
- SLPT fails in cases where the data is **non-linearly separable (NLS)**, which restricts its application to many real-world classification problems.

**1.2. Recursive Deterministic Perceptron (RDP)**
- The **Recursive Deterministic Perceptron (RDP)** is a **multilayer generalization** of SLPT.
  - Unlike SLPT, the RDP can solve any two-class classification problem, even for NLS sets.
- The RDP network works by progressively adding **intermediate neurons (INs)** that correspond to SLPTs.
  - These intermediate neurons help to **augment the affine dimension** of input vectors.
  - The output of each IN is added as a new component to the input vector, transforming the NLS problem into a linearly separable (LS) one.

**1.3. Key Idea Behind the RDP Model**
- The construction of an RDP model is **automatic** and guarantees **convergence to a solution**.
  - This ensures that the RDP can always find a decision boundary for any two-class classification problem.
  - The transformation is done step by step, progressively adding neurons until the two classes become linearly separable.
  
**1.4. Proposed Growing Methods**
- The authors propose **three growing methods** for constructing RDPs:
  - **Batch Learning**: A method that builds the RDP in one step using the entire data set.
  - **Incremental Learning**: Dynamically updates the network when new patterns are introduced.
  - **Modular Learning**: Decomposes the problem into smaller, independently solvable sub-problems and later combines the solutions.

**1.5. Knowledge Extraction from RDP Networks**
- RDP networks allow for easy extraction of embedded knowledge.
  - The decision region of an RDP can be expressed as a finite union of open polytopes (geometrical regions).
  - This allows for **rule extraction** from the model, making it more transparent than black-box models like backpropagation (BP).

**1.6. Comparison with Other Methods**
- The paper contrasts the RDP with other growing models, including **Cascade Correlation (CC)** and **backpropagation (BP)**.
  - Unlike CC and BP, the RDP guarantees convergence and has a more explicit representation of decision regions.
  - The geometrical properties of the RDP model are highlighted as an advantage over these models, particularly in knowledge extraction and interpretability.

**1.7. XOR Problem as an Example**
- The **Exclusive-OR (XOR)** problem is used to illustrate the RDP's capability:
  - The XOR problem is NLS, meaning it cannot be solved by SLPT.
  - The RDP model solves it by augmenting the input space using intermediate neurons (IN1 and IN2).
  - A two-layer RDP network successfully separates the XOR classes, providing a solution that is guaranteed to converge.

## **3. Growing Methods for Constructing Recursive Deterministic Perceptron (RDP) Models**

In this section, the authors propose three methods for building RDP models: Batch Learning, Incremental Learning, and Modular Learning. Each method presents a different approach to construct RDP models while handling classification problems, including non-linearly separable (NLS) problems.



### **3.1 Batch Learning Method**

**3.1.1. Overview of the Batch Method**
- The batch method builds an RDP by selecting linearly separable (LS) subsets from the input data.
- The algorithm aims to construct an RDP by finding a maximal LS subset of points that belong to the same class at each step.
- The process continues iteratively, adding intermediate neurons (INs) to transform non-linearly separable sets into linearly separable ones.

**3.1.2. Algorithm Description**
- The batch learning algorithm starts by considering all input points and at each step selects a linearly separable subset.
- It adds intermediate neurons to the RDP based on the hyperplanes separating selected subsets.
- The process terminates when the classes become linearly separable, with the RDP being constructed from these intermediate neurons.

**3.1.3. Example of Batch Learning**
- The authors provide an example involving two classes A and B in a two-dimensional classification problem.
- The batch learning method is applied to select subsets of points that are linearly separable and adds neurons to the RDP model until the problem becomes solvable.
  
**3.1.4. Discussion of Batch Method**
- **Advantages**: This method is useful for static problems where the input data does not change over time.
- **Efficiency**: In the worst-case scenario, the batch method constructs an RDP with at most \( n-1 \) intermediate neurons, where \( n \) is the number of input patterns.
- **Challenges**: Optimizing the number of steps, and thereby minimizing the number of neurons, is proven to be NP-complete in cases of arbitrary dimensions.




### **3.2 Incremental Learning Method**

**3.2.1. Overview of the Incremental Method**
- This method allows for incremental or progressive learning by adding new knowledge to the RDP model without disturbing the previously acquired information.
- It is designed to handle dynamic problems where new data points are introduced after the initial model is built.

**3.2.2. Algorithm Process**
- The incremental method modifies the existing RDP by adding a new intermediate neuron or adjusting an existing one if a new data point is not correctly classified.
- The advantage of this method is that it preserves the knowledge embedded in the RDP while incorporating new patterns, avoiding catastrophic interference common in other models like backpropagation (BP).

**3.2.3. Biological Soundness**
- This method is considered more biologically plausible because it allows the network to evolve as new data is presented, similar to how humans learn over time.

**3.2.4. Discussion of Incremental Learning**
- **Efficiency**: If applied throughout the entire learning process, the RDP constructed via the incremental method contains at most \( \frac{n}{2} \) intermediate neurons, where \( n \) is the number of input patterns.




### **3.3 Modular Learning Method**

**3.3.1. Overview of the Modular Method**
- The modular method decomposes the problem into smaller, independently solvable sub-problems, then combines the results to construct a complete RDP.
- This approach can also be used to combine existing RDP models.

**3.3.2. Process**
- The method allows for parallel learning of sub-problems, which can be combined into a single RDP without retraining.
- It is useful for handling complex problems that can be broken down into simpler components.

**3.3.3. Complexity**
- The modular method may result in a topology containing up to \( O(n^2) \) intermediate neurons in the worst case. However, if the learning process is parallelized, the time complexity can be reduced to \( O(\log(n)) \).

**3.3.4. Discussion**
- This method is ideal for "dynamic" problems where sub-problems need to be combined or split as needed.
- It offers flexibility and scalability in constructing RDPs for more complex tasks.



## **4. Knowledge Extraction from an RDP Neural Network**

This section delves into how the Recursive Deterministic Perceptron (RDP) model allows for transparent knowledge extraction. Unlike traditional feedforward networks, such as backpropagation (BP), which are often regarded as "black boxes," the RDP’s geometrical properties make it easier to extract and interpret embedded knowledge.



### **4.1 Decision Region**

**4.1.1. Difficulty of Knowledge Extraction in Other Models**
- Feedforward neural networks (e.g., BP models) present a challenge when it comes to explaining their decision-making process.
  - These networks often function as "black boxes," making it difficult to understand how they reach specific conclusions.
  - Although extensive research has been conducted in this area, effective methods for extracting knowledge from neural networks are still under development.

**4.1.2. RDP's Transparent Knowledge Representation**
- RDP models, due to their geometric structure, allow knowledge to be expressed as a **finite union of open polytopes** in \( \mathbb{R}^d \).
  - This means that the decision region of an RDP can be represented using simple geometric shapes.
  - A decision region refers to the part of the input space where the RDP assigns the same classification label.

**4.1.3. Definition of Open Polytope**
- A subset \( S \) of \( \mathbb{R}^d \) is defined as an open polytope if it can be expressed as an intersection of hyperplanes.
  - Each hyperplane is associated with one of the neurons in the RDP model.
  - The decision region of the RDP is therefore a finite combination of these polytopes, making the network’s decision boundaries explicit.

**4.1.4. Proposition 21**
- The decision region of an RDP \( P \) on \( \mathbb{R}^d \) is always a finite union of open polytopes.
  - For an RDP with height 1, the decision region corresponds to a half-space.
  - For higher dimensions, the decision region is built recursively from the intermediate neurons, preserving the property of being a union of polytopes.



### **4.2 Boolean Operations on Decision Regions**

**4.2.1. Combining Decision Regions Using Boolean Operations**
- It is possible to combine the decision regions of different RDPs using **Boolean operations** (e.g., union, intersection).
  - These operations allow for constructing more complex decision regions by combining simpler ones.
  
**4.2.2. Theorem 22**
- **Theorem 22** demonstrates how to construct new RDPs from existing ones by applying Boolean operations.
  - For instance:
    - \( P_3 \) can be constructed to represent the union of two decision regions.
    - \( P_4 \) can represent the intersection of two decision regions.
    - \( P_5 \) can represent the complement of a decision region.
  - These Boolean operations enable more versatile and interpretable decision-making models using RDPs.



### **4.3 Example of XOR Problem**

**4.3.1. XOR Decision Region**
- The XOR problem is used to illustrate how decision regions are formed in an RDP model.
  - The decision region of the RDP corresponding to the XOR problem is shown as a combination of two half-spaces.
  - The decision boundary, represented by the union of two open polytopes, clearly demonstrates the network's decision-making process.

**4.3.2. Visualization of the XOR Decision Region**
- A graphical representation of the decision region for the XOR problem is provided, showing the two regions separated by hyperplanes.
  - This highlights the transparency of the RDP model in representing complex classification problems with explicit decision boundaries.

##**5. Discussion and Concluding Remarks**



###**5.1 Overview of the RDP Model**
- The **Recursive Deterministic Perceptron (RDP)** is introduced as a neural network model that can solve both linearly separable (LS) and non-linearly separable (NLS) classification problems.
- The RDP model generalizes the Single Layer Perceptron Topology (SLPT), enhancing its capabilities for handling NLS data.
  


###**5.2 Summary of Growing Methods**
- The paper presented **three growing methods** to construct RDPs:
  1. **Batch Learning**: Suitable for static problems, where the entire dataset is available upfront.
  2. **Incremental Learning**: Best for dynamic problems, allowing new knowledge to be incorporated without disturbing previously learned information.
  3. **Modular Learning**: This method is highly useful for breaking down complex problems into sub-problems, solving them independently, and combining them.



###**5.3 Key Results**
- The three proposed methods are capable of constructing RDP models to solve LS and NLS problems efficiently.
- For batch learning, the worst-case scenario produces a model with \( n-1 \) intermediate neurons (INs), where \( n \) is the number of patterns.
- Incremental learning leads to models with \( n/2 \) INs in the worst case, while modular learning produces RDPs with \( O(n^2) \) neurons, which can be reduced to \( O(\log(n)) \) with parallel construction.



###**5.4 Knowledge Extraction**
- One of the key advantages of the RDP model is the ability to extract knowledge from the network.
  - Knowledge embedded in RDP models can be represented as **finite unions of open polytopes**, making it easier to interpret the decision boundaries.
  - This transparency in knowledge extraction gives RDP a significant advantage over traditional feedforward neural networks like **Backpropagation (BP)**, which are often seen as "black boxes."



###**5.5 Comparison with Other Methods**
- RDP models show clear benefits in terms of interpretability and transparency.
  - **BP models** face difficulties in knowledge extraction, while RDP’s structure allows for transparent decision-making through geometrically defined decision regions.
- The ability to use **Boolean operations** to combine decision regions of RDPs further increases the versatility of the RDP approach.



###**5.6 Future Work**
- The authors suggest further research into optimizing the search strategies for **linear separability (LS) subsets**, which will improve the construction of minimal RDP topologies.
- Further study is also encouraged in exploring applications for **image compression**, and the **topological and geometrical characterization** of separability classes for RDP models.
