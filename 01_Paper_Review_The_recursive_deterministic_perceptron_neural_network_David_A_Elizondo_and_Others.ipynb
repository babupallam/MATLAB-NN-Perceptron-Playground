**Title**: The Recursive Deterministic Perceptron Neural Network

**Authors**: Mohamed Tajine and David Elizondo

**Published**: Neural Networks, 1998

## Aim of the Paper:
The paper introduces a new feedforward multilayer neural network model, the **Recursive Deterministic Perceptron (RDP)**, which generalizes the single-layer perceptron topology. The goal is to address the limitations of the single-layer perceptron (SLP), particularly its inability to solve non-linearly separable classification problems.

## Key Proposals:
1. **Recursive Deterministic Perceptron (RDP)**: A neural network model designed to solve any two-class classification problem, even when the two sets are not linearly separable.
   
2. **Growing Method**: The RDP is built by successively adding intermediate neurons (INs), each corresponding to a single-layer perceptron. These neurons augment the input vector space, transforming non-linearly separable problems into linearly separable ones.

3. **Generalization for Multi-class Classification**: The paper extends the RDP model for the classification of more than two classes (m classes) by introducing a new notion of linear separability for multiple classes.

## Contributions:
- The paper presents several construction algorithms for building RDPs by progressively adding neurons. The proposed RDP is shown to be capable of separating m classes and can approximate continuous functions.
- Comparative results on classification accuracy and network topology size are provided against other neural network methods like backpropagation (BP) and cascade correlation (CC).



# Detailed Review

## **1. Introduction**



### **1.1. Limitations of the Single Layer Perceptron (SLP)**
- The Single Layer Perceptron (SLP) can only solve linearly separable classification problems.
- Classification problems dealing with non-linearly separable sets (NLS) cannot be handled by SLP.
  - Example: XOR problem, where two classes are not linearly separable in their original input space.



### **1.2. Recursive Deterministic Perceptron (RDP)**
- Proposed as a **multilayer generalization** of the SLP.
- Capable of always separating two classes, even if they are non-linearly separable (NLS).
  


### **1.3. Key Idea Behind RDP**
- The RDP incrementally increases the dimension of the input vector by using the outputs of a sequence of **intermediate neurons (INs)**.
- Each IN corresponds to a simple SLP.
- The aim is to transform a non-linearly separable (NLS) problem into a **linearly separable (LS)** one by:
  - Adding new components to the input vectors.
  - Gradually transforming the problem through a series of transformations.



### **1.4. Working Mechanism of RDP**
- In each time step, a new IN is added if the classes remain non-linearly separable.
  - The IN is constructed by selecting a subset of points that are **linearly separable** (LS) from the rest.
  - This continues until the classes become LS.
- **Guaranteed Termination**: The algorithm stops after at most **n-1 steps**, where "n" is the number of input patterns.
  - The final network is a composition of several SLPs and forms a **multilayer perceptron** topology.

### **1.6. Comparison with Other Growing Neural Networks**
- Several growing algorithms have been developed for constructing neural networks by adding neurons during training.
  - **Cascade Correlation (CC)**: Adds new intermediate neurons (INs) during learning, but uses backpropagation (BP) for training.
    - BP does not guarantee convergence.
  - **Other Models**:
    - **Gallant’s Perceptron Algorithm**: Selects linearly separable subsets but has no guaranteed convergence.
    - **Geometrical Methods**: Other methods like Bose and Garga’s approach use convex hulls and Voronoi diagrams, but are often inefficient in terms of the number of neurons.
    - **Expanding-and-Truncate Learning (ETL)**: A method with guaranteed convergence but limited to binary classification.

### **5.4. Advantages and Disadvantages of RDP**

- **Advantages**:
  - **Self-Determined Topology**: RDP determines the number of intermediate neurons automatically, unlike BP where the number of hidden units must be predefined.
  - **Guaranteed Convergence**: The RDP is guaranteed to converge, as proven in several theorems, unlike BP or CC methods, which may get stuck in local minima.
  - **Avoidance of Catastrophic Interference**: The RDP's incremental learning method prevents catastrophic forgetting of previously learned information.
  - **Rule Extraction**: RDP can simplify rule extraction since its decision regions are well-defined by hyperplanes, making the neural network more interpretable.

- **Disadvantages**:
  - **Fan-Ins**: The RDP’s architecture may involve a high number of connections, which could restrict hardware implementations. Optimizing this is suggested but was not the focus of the paper.
  - **Lower Generalization in Some Cases**: RDP's generalization level was slightly lower compared to highly optimized methods like RuleNeg and Rulex.
  - **Computational Time**: Finding the linear separable (LS) subsets to build the intermediate neurons can be computationally expensive.

## **6. Discussion and Concluding Remarks**



### **6.1. Overview of the RDP Model**
- The Recursive Deterministic Perceptron (RDP) was introduced as a generalization of the single-layer perceptron topology (SLPT) capable of solving non-linearly separable (NLS) two-class problems.
- The RDP achieves this by augmenting the input vector’s dimensions through Intermediate Neurons (INs).
  - These neurons function similarly to hidden units in backpropagation (BP) algorithms by transforming NLS problems into linearly separable (LS) ones.



### **6.2. RDP's Algorithmic Strategy**
- The method of selecting the subsets of linearly separable points (LS subsets) was based on maximizing the cardinality of these subsets. However, this strategy has not been proven to yield optimal or minimal network topologies.
- **Complexity**: The selection strategy for LS subsets becomes NP-complete when the dimensionality of the input vector is arbitrary.
- The authors call for further research to develop better search strategies to optimize the RDP topology.



### **6.3. Generalization to m-Class Problems**
- The paper also proposed a new notion of **linear separability for m classes**, where m classes are separated by a single hyperplane translated \(m+1\) times.
- To solve multi-class classification problems, the original m-class problem is reduced to a series of m! two-class problems.



### **6.4. Evaluation of Results**
- The proposed algorithms were applied to six classification problems, and the results were compared with other neural network methods, including BP, Cascade Correlation (CC), RuleNeg, and Rulex.
- **Results**: The RDP performed comparably to these methods in terms of the percentage of test data correctly classified, though in some cases, methods like RuleNeg and Rulex had fewer intermediate neurons.



### **6.5. Advantages of the RDP Model**
- **Topology Adaptation**: Unlike other models, the RDP can adapt its topology by determining the number of intermediate neurons automatically.
- **Guaranteed Convergence**: The RDP avoids local minima, a problem common in methods like BP.
- **Rule Extraction**: RDPs provide an easier method for rule extraction as their decision regions are defined by hyperplanes.



### **6.6. Disadvantages of the RDP Model**
- **Fan-Ins**: The network may have numerous connections (fan-ins), which can complicate hardware implementations. This can be optimized by minimizing connections with zero-weight components.
- **Topology Size**: The RDP often requires more neurons than optimized methods like RuleNeg or Rulex.



### **6.7. Future Research**
- The authors recommend further research to optimize the selection strategy of LS subsets to improve the efficiency of the RDP’s topology.
- Other areas for exploration include:
  - Image compression applications.
  - Topological and geometrical characterization of separability classes for RDP.
  - Exploring the lower bounds for the height and width of RDP networks.
